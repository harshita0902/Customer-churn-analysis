```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/WA_Fn-UseC_-Telco-Customer-Churn.csv")
```

```python
df['TotalCharges'] = df['TotalCharges'].replace(' ', np.nan)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'])
```

```python
df['TotalCharges'].dtype
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.describe(include='all')
```

```python
df['Churn'].value_counts()
df['Churn'].value_counts(normalize=True) * 100
```

```python
sns.countplot(x='Churn', data=df)

plt.title("Customer Churn Distribution")

plt.show()
```

### Insight:
Most customers stayed with the company, while a smaller percentage churned. The dataset is moderately imbalanced.

```python
sns.boxplot(x='Churn', y='tenure', data=df)

plt.title("Tenure vs Churn")

plt.show()
```

```python
sns.histplot(data=df, x='tenure', hue='Churn',multiple='stack')
plt.title("Tenure Distribution by Churn")
plt.show()
```

Customers with low tenure are much more likely to churn, especially within the first few months.


```python
sns.countplot(x='Contract', hue='Churn', data=df)

plt.title("Contract Type vs Churn")

plt.xticks(rotation=15)

plt.show()
```

Month-to-month customers show significantly higher churn compared to one-year and two-year contract customers.

```python
sns.countplot(x='InternetService', hue='Churn', data=df)

plt.title("Internet Service vs Churn")

plt.show()
```

Customers with Fibre Optic churned the most.

```python
sns.countplot(x='TechSupport', hue='Churn', data=df)

plt.title("Tech Support vs Churn")

plt.show()
```

### Business Insight:
Customers without tech support churn significantly more, suggesting that support services improve customer retention.
